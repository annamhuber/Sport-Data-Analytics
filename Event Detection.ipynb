```python
import pandas as pd
import numpy as np
import matplotlib
```

```python
df_badminton = pd.read_csv('Data/Badminton.tsv', sep='\t', skiprows=[1])
df_badminton= df_badminton.applymap(lambda x: x.replace(",",".")).astype("float64")
```

```python
df_badminton[["Ankle, right X","Ankle, right Y","Ankle, right Z"]].plot()
```
